```python
import random

wantToPlay = True  #variable to now if user want to play again
wins = 0    #wins count
loses = 0   #loses count
print('Welcome! What is your name?')  
name =  input()

while wantToPlay:  #while loop will iterate until player giveup
    print()        #empty line
    print('Hi '+name+'! Welcome to the "Guess the Number" game')
    print('Please select the difficulty level: ')
    print('EASY')
    print('MODERATE')
    print("HARD")
    print()
   
    while True:    #while loop to select difficulty
        difficulty = input().upper()  
        # using input function to get input from user and upper function to change user input value to uppercase
        if difficulty == 'EASY':      #if condition to find out which difficulty player selected
            randomNumber = random.randint(1,10) #random number generator
            print()
            print('Guess the number between 1 to 10')
            for i in range(5):   #for loop that will iterate for 5 times for easy difficulty
                guess = input()  #input function to take user guess number
                guess = int(guess)
                if guess>randomNumber:
                        print('your guess is too high')
                if guess<randomNumber:
                        print('your guess is too low')
                if guess == randomNumber: #condition to check the entered value is correct or not
                    print('YOU WIN!')
                    wins+=1
                    break

                if i==4:
                    print('YOU LOOSE!')
                    loses+=1
            break
         # same logic for moderate difficulty           
        elif difficulty == 'MODERATE':
            randomNumber = random.randint(1,100)
            print()
            print('Guess the number between 1 to 100')
            for i in range(8):
                guess = input()
                guess = int(guess)
                if guess>randomNumber:
                        print('your guess is too high')
                if guess<randomNumber:
                        print('your guess is too low')
                if guess == randomNumber:
                    print('YOU WIN!')
                    wins+=1
                    break
                
                if i==7:
                    print('YOU LOOSE!')
                    loses+=1
            break
        # same logic for hard difficulty            
        elif difficulty == 'HARD':
            randomNumber = random.randint(1,1000)
            print()
            print('Guess the number between 1 to 1000')
            for i in range(10):
                guess = input()
                guess = int(guess)
                if guess>randomNumber:
                        print('your guess is too high')
                if guess<randomNumber:
                        print('your guess is too low')
                if guess == randomNumber:
                    print('YOU WIN!')
                    wins+=1
                    break
                
                if i==9:
                    print('YOU LOOSE!')
                    loses+=1
            break
        # if user enter wrong difficulty level then it will ask again to enter correct difficulty level            
        else:
            print('Please enter EASY or MODERATE or HARD only')
                
    print()
    # prints player name ,wins, losses
    print(name+'s Score: '+str(wins)+' wins, '+str(loses)+' loses') #prints player name ,wins, losses
    # prints player winning percentage
    print('Enter YES to play again or No to quit')
    while True:
        choice = input().upper()
        if choice == 'NO':
            wantToPlay =False
            break
            
        elif choice == 'YES':
            wantToPlay = True
            break
            
        else:
            print ('Please enter YES or NO only')
print('Thanks for playing ' +name+'!')

```
