```python
#!/usr/bin/env python3

#6 - 1

# Welcome message
def display_welcome():
    print("The Test Scores program")
    print("Enter 'x' to exit")
    print("")

# Get scores from user                
def get_scores():
    # initialize list
    scoreslist = []

    # loop to get scores
    while True:
        # print score number input
        score = input("Enter test score: ")

        # set condition to break loop
        if score == "x" or score == "X":
            break
        # convert test score from string to integer
        scoreslist.append(int(score))
    # return list
    return scoreslist

# Process scores
def process_scores(scoreslist): # add parameter
    # initialize variables
    # len() function returns the number of items in an object
    n = len(scoreslist)
    total = 0
    # loop to get total
    for score in scoreslist:
        total += score
    
    numberofscores = len(scoreslist)
    average = round(total / n)
    scoreslist.sort()
    lowest = scoreslist[0]
    highest = scoreslist[-1]
    median = 0

    # calculate median
    if n % 2 == 0:
        median = (scoreslist[n//2] + scoreslist[n//2 - 1]) / 2
    else:
        median = scoreslist[n//2]

    # format and display the result  
    print()
    print("Total:              ", total)
    print("Number of Scores:   ", numberofscores)
    print("Average score:      ", average)
    print("Lowest score:       ", lowest)
    print("Highest score:      ", highest)
    print("Median score:       ", median)

# Main function, calls other functions
def main():
    display_welcome()
    scoreslist = get_scores()
    process_scores(scoreslist)

# if started as the main module, call the main function
if __name__ == "__main__":
    main()
```

```python
#!/usr/bin/env python3

#6 - 2

# Welcome message
def display_menu():
    print("COMMAND MENU")
    print("list - List all movies")
    print("add -  Add a movie")
    print("del -  Delete a movie")
    print("find - Find a movie by year")
    print("exit - Exit program")
    print()    

# List movies
def list(movie_list):
    # for loop to print movies
    for i, movie in enumerate(movie_list):
        # list all 3 items in movie list
        name, year, price = movie
        # print movie number, name, year, and price
        print(f"{i+1}. {name} ({year}) @ {price:.2f}")
    print()

# Add a movie
def add(movie_list):
    name = input("Name: ")
    year = input("Year: ")
    price = float(input("Price: "))
    # create list for movie
    movie = []
    # append movie list with name, year, and price
    movie.append(name)
    movie.append(year)
    movie.append(price)
    movie_list.append(movie)
    print(movie[0] + " was added.\n")

# Delete a movie    
def delete(movie_list):
    # convert input to integer
    number = int(input("Number: "))
    # set condition for invalid input
    if number < 1 or number > len(movie_list):
        print("Invalid movie number.\n")
    # delete movie
    else:
        movie = movie_list.pop(number-1)
        print(movie[0] + " was deleted.\n")

# Find a movie by year
def find(movie_list):
    # convert input to integer
    year = int(input("Year: "))
    # for loop to find movie
    for movie in movie_list:
        # set condition for movie found
        if movie[1] == year:
            print(movie[0] + " was released in " + str(year) + ".\n")
            return
    print("No movies were released in " + str(year) + ".\n")
    

# Main function, calls other functions
def main():
    movie_list = [["Monty Python and the Holy Grail", 1975, 9.95],
                  ["On the Waterfront", 1954, 5.59],
                  ["Cat on a Hot Tin Roof", 1958, 7.95]]

    # display welcome message and command menu
    display_menu()
    while True:        
        command = input("Command: ")
        if command == "list":
            list(movie_list)
        elif command == "add":
            add(movie_list)
        elif command == "del":
            delete(movie_list)
        elif command == "find":
            find(movie_list)
        elif command == "exit":
            break
        else:
            print("Not a valid command. Please try again.\n")
    print("Bye!")

# if started as the main module, call the main function
if __name__ == "__main__":
    main()
```
