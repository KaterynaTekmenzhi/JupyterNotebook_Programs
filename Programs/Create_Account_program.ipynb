```python
#!/usr/bin/env python3

# Movie Catalog program 12 - 2

# The movie program will allow the user to maintain a list of movies. Each movie has a name and a year.
# The program will allow the user to add movies, list all movies, and delete movies. 
# The program will also allow the user to save the movie list to a text file and to load the movie list from a text file.

# The program will display a menu of options. The user will be able to choose from the following menu options:
def display_menu():
    print("The Movie List program")
    print("COMMAND MENU")
    print("list - List all movies")
    print("add - Add a movie")
    print("del - Delete a movie")
    print("exit - Exit program")
    print()

# The program will store the movie list in a list of dictionaries. Each dictionary will represent a movie.
def list(movie_list):
    if len(movie_list) == 0:
        print("There are no movies in the list.\n")
    else:
        i = 1
        for movie in movie_list:
            print(f"{i}. {movie['name']} ({movie['year']})") # print dictionary, name key, year key
            i += 1
        print()

# The program will allow the user to add a movie. The user will enter the name and year of the movie.
def add(movie_list):
    name = input("Name: ")
    year = int(input("Year: ")) 
    movie = {"name": name, "year": year} # create dictionary
    movie_list.append(movie) # add dictionary to list
    print(f"{name} was added.\n")

# The program will allow the user to delete a movie. The user will enter the number of the movie to delete.
def delete(movie_list):
    number = int(input("Number: "))
    if number < 1 or number > len(movie_list):
        print("Invalid movie number.\n")
    else:
        movie = movie_list.pop(number-1)
        print(f"{movie['name']} was deleted.\n") # print dictionary, name key

# The main function will display the menu and respond to the userâ€™s menu choice until the user chooses to exit the program.
def main():
    # dictionary list
    movie_list = [
        {"name": "Monty Python and the Holy Grail", "year": 1975},
        {"name": "On the Waterfront", "year": 1954},
        {"name": "Cat on a Hot Tin Roof", "year": 1958}
    ]
    display_menu()

    # loop to get and process user input
    while True:
        command = input("Command: ")
        if command == "list":
            list(movie_list)
        elif command == "add":
            add(movie_list)
        elif command == "del":
            delete(movie_list)
        elif command == "exit":
            break
        else:
            print("Not a valid command. Please try again.\n")
    print("Thanks for using the Movie List program!")
    print("Bye!")

# call main function
if __name__ == "__main__":
    main()
```
