```python
# Kateryna Tekmenzhi - 11/10/2023 - Code Practice 7 - Data333

#!/usr/bin/env python3

# Book Catalog Program  12 - 1

# This program allows users to add books to a catalog of books, list all books in the catalog, and delete books from the catalog.
# Each book entry has a title, author name, and publication year. The program uses a dictionary to store the book information.

# Show book info - prompt for title and display the author name and publication year
def show_book(book_catalog):
    title = input("Title: ")
    if title in book_catalog:
        book = book_catalog[title]
        print("Title:    " + title)
        print("Author:   " + book["author"])
        print("Pub year: " + book["pubyear"])
    else:
        print("Sorry, " + title + " doesn't exist in the catalog.")

# list all the books in the catalog - attach associated author and publication year, and print them out
def show_all_books(book_catalog):
    for title in book_catalog:
        book = book_catalog[title]
        print("Title:    " + title)
        print("Author:   " + book["author"])
        print("Pub year: " + book["pubyear"])
        print()

# add a book to the catalog - prompt for title, author name, and publication year
def add_edit_book(book_catalog, mode):
    title = input("Title: ")
    # Check if the book exists in the catalog
    if mode == "add" and title in book_catalog:
        print(title + " already exists in the catalog.")
        response = input ("Would you like to edit it? (y/n): ").lower()
        # If the book exists, ask the user if they want to edit it
        if(response != "y"):
            return
        # If the user wants to edit the book, delete the book from the catalog
    elif mode == "edit" and title not in book_catalog:
        print(title + " doesn't exist in the catalog.")
        response = input("Would you like to add it? (y/n): ").lower()
        if (response != "y"):
            return
        
    # Prompt the user for the book data
    author = input("Author name: ")
    pubyear = input("Publication year: ")
    
    # Create a dictionary for the book data
    book = {"author": author,
            "pubyear": pubyear}

    # Add the book data to the catalog using title as key
    book_catalog[title] = book

    # Display a confirmation message
    print(title + " was added to the catalog.")

# delete a book from the catalog - prompt for title and delete the book from the catalog
def delete_book(book_catalog):
    title = input("Title: ")
    # Check if the book exists in the catalog
    if title in book_catalog:
        # If the book exists, delete it from the catalog
        del book_catalog[title]
        print(title + " removed from catalog.")
    else:
        print(title + " doesn't exist in the catalog.")

# display the menu, prompt for a command, and perform the requested action
def display_menu():
    print("The Book Catalog program")
    print()
    print("COMMAND MENU")
    print("show - Show book info")
    print("list - List all books")
    print("add -  Add book")
    print("edit - Edit book")
    print("del -  Delete book")
    print("exit - Exit program")

# main function, which calls the display_menu function and then prompts for a command
def main():
    display_menu()
    book_catalog = {
        "Moby Dick": 
            {"author" : "Herman Melville",
             "pubyear" : "1851"},
        "The Hobbit":
            {"author" : "J. R. R. Tolkien", 
             "pubyear" : "1937"},
        "Slaughterhouse Five":
            {"author" : "Kurt Vonnegut",
             "pubyear" : "1969"}
        }
    
    # User input is taken to determine the command. Depending on the command, the appropriate function is called.
    while True:
        print()
        command = input("Command: ").lower()
        if command == "show":
            show_book(book_catalog)
        elif command == "add":
            add_edit_book(book_catalog, mode="add")
        elif command == "list":
            show_all_books(book_catalog)
        elif command == "edit":
            add_edit_book(book_catalog, mode="edit")
        elif command == "del":
            delete_book(book_catalog)
        elif command == "exit":
            print("Thank you for using the Book Catalog Program")
            print("Bye!")
            break
        else:
            print("Unknown command. Please try again.")

# Call the main function
if __name__ == "__main__":
    main()
```
