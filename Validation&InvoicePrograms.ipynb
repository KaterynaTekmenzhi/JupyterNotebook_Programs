```python
# 9 - 1
#!/usr/bin/env python3

from decimal import Decimal
from decimal import ROUND_HALF_UP

# display a proffessional title
print("The Invoice program")
print("Use integer or decimal numbers for all entries.\n")

# initialize variables, set loop condition to y so it runs at least once. 
choice = "y"
# Modify the program so that it continues only if the user enters y or Y.
# If the user enters anything else, terminate the program.
while choice == "y":
    
    # get user entry
    order_total = Decimal(input("Enter order total:     "))
    order_total = order_total.quantize(Decimal("1.00"), ROUND_HALF_UP)
    print()               

    # determine discount percent
    if order_total > 0 and order_total < 100:
        discount_percent = Decimal("0")
    elif order_total >= 100 and order_total < 250:
        discount_percent = Decimal(".1")
    elif order_total >= 250:
        discount_percent = Decimal(".2")

    # calculate results
    # Modify the program so that it displays the order total and invoice total as U.S. currency values 
    # with dollar signs and two decimal points.

    discount = order_total * discount_percent
    discount = discount.quantize(Decimal("1.00"), ROUND_HALF_UP)
    subtotal = order_total - discount
    sales_tax_percent = Decimal(".05")
    sales_tax = sales_tax_percent * subtotal
    sales_tax = sales_tax.quantize(Decimal("1.00"), ROUND_HALF_UP)
    invoice_total = subtotal + sales_tax

    # Add a shipping cost which should be 8.5% (.085) of the subtotal. Use the decimal module to 
    # make sure each monetary value has the correct number of decimal places. 
    # Display the shipping cost on the invoice.
    shipping_cost_percent = Decimal(".085")
    shipping_cost = shipping_cost_percent * subtotal
    shipping_cost = shipping_cost.quantize(Decimal("1.00"), ROUND_HALF_UP)
    invoice_total = subtotal + sales_tax + shipping_cost

    # order total, invoice total are displayed with dollar signs, commas, and two decimal points.
    order_total = "${:,.2f}".format(order_total)
    invoice_total = "${:,.2f}".format(invoice_total)

    # discount, subtotal, and sales tax are displayed with two decimal points, but without dollar signs, commas, or percent signs.
    discount = "{:.2f}".format(discount)
    subtotal = "{:.2f}".format(subtotal)
    sales_tax = "{:.2f}".format(sales_tax)

    # shipping cost are be displayed with two decimal points, but without dollar signs or commas.
    shipping_cost = "{:.2f}".format(shipping_cost)

    # display the results
    print("Order total:      ", order_total)
    print("Discount amount:  ", discount)
    print("Subtotal:         ", subtotal)
    print("Shipping cost:    ", shipping_cost)
    print("Sales tax:        ", sales_tax)
    print("Invoice total:    ", invoice_total)
    print()
    choice = input("Continue? (y/n): ")    
    print()
    
print("Thank you for using the Invoice program.")
print("Have a great day!\n")
```

```python
# 10 - 1
#!/usr/bin/env python3

# display a proffessional title
print("Create Customer File program")
print()
print("Welcome new customer!")


def main():
    # get user's first and last name
    full_name = get_full_name()
    print()
    # get user's email address
    email = get_email()
    print()
    # display a message including the phone number format that uses dots to group the digits.
    # get user's phone number, group digits with dots
    phone_number = get_phone_number()
    phone_number = phone_number[:3] + "." + phone_number[3:6] + "." + phone_number[6:]
    print()
    # get user's password
    password = get_password()
    print()    
    # display a welcome message that includes the user's first name.
    first_name = get_first_name(full_name)   
    print("Hi " + first_name + ", thanks for creating an account.") 
    # display a confirmation message that includes the user's full name, phone number, and email address.
    print("We'll text your confirmation code to this number:" + phone_number)
    print("We'll send your monthly invoice to this email address:" + email)
               
# Create and use a function to get a valid full name.
def get_full_name():
    while True:
        name = input("Enter full name:       ").strip()
        if " " in name:
            return name
        else:
            print("You must enter your full name.")

# Create and use a function to get the first name from the full name.
def get_first_name(full_name):
    index1 = full_name.find(" ")
    first_name = full_name[:index1]
    return first_name

# Create and use a function to get a valid password.
def get_password():
    while True:
        digit = False
        cap_letter = False
        password = input("Enter password:        ").strip()
        for char in password:
            if char.isdigit():
                digit = True
            elif char.isupper():
                cap_letter = True
        if digit == False or cap_letter == False or len(password) < 8:
            print("Password must be 8 characters or more \n" +
                  "with at least one digit and one uppercase letter.")
        else:
            return password
        
# Create and use a function to get a valid phone number. 
# To do that, remove all spaces, dashes, parentheses, and periods from the number. 
# Then, check to make sure that the phone number consists of 10 characters that are digits.
def get_phone_number():
    while True:
        phone_number = input("Enter phone number:    ").strip()
        phone_number = phone_number.replace(" ", "")
        phone_number = phone_number.replace("-", "")
        phone_number = phone_number.replace("(", "")
        phone_number = phone_number.replace(")", "")
        phone_number = phone_number.replace(".", "")
        if phone_number.isdigit() and len(phone_number) == 10:
            return phone_number
        else:
            print("Please enter a valid 10 digit phone number.")
        
# Create and use a function to get a valid email address. 
# To be valid, the address has to contain an @ sign and end with ".com"
def get_email():
    while True:
        email = input("Enter email address:   ").strip()
        if "@" in email and email.endswith(".com"):
            return email
        else:
            print("Please enter a valid email address.")

# Call the main function.   
if __name__ == "__main__":
    main()
```
