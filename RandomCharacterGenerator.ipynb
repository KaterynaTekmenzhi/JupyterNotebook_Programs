```python
# Scenario
# Create a character generator program for a Role Playing game. The character can be any type of character you want it to be. You can opt for standard Dungeons and Dragons style here, or go with any other character you might want. 

# The program must do the following:
# 1. Ask if the user would like to create a character. If the user answers yes, then the program should continue. If the user answers no, then the program should end.
# 2. Ask for a character name. Or provide a name for them. 
# 3. import and use the random module
# 4. generate at least 5 characteristics of a character (such as strength, intellect, hit points, wisdom, charisma, etc.) and assign a random value to each of them. 
# 5. Use at least 2 functions in your program. One of the functions must pass a parameter.
# 6. Use a random number and if statement to generate a class/race or character type (such as human warrior, or alien from Mars, etc.) 
# 7. Output the character information to the user in a clear format.
# 8. Ask the user if they would like to create another character. If yes, then repeat the process. If no, then end the program.

# import the random module
import random

# define a function to generate a random number
def main():

    # define a function to generate a random character type
    def random_character_type():
        # generate a random number between 1 and 3
        number = random.randint(1, 8)
        # if the number is 1, then return "lawyer"
        if number == 1:
            return "Lawyer"
        # if the number is 2, then return "Queen of England"
        elif number == 2:
            return "Queen of England"
        # if the number is 3, then return "Delivery Man"
        elif number == 3:
            return "Delivery man"
        # if the number is 4, then return "Doctor"
        elif number == 4:
            return "Doctor"
        # if the number is 5, then return "Teacher"
        elif number == 5:
            return "Teacher"
        # if the number is 6, then return "Student"
        elif number == 6:
            return "Student"
        # if the number is 7, then return "Police Officer"
        elif number == 7:
            return "Police Officer"
        # if the number is 8, then return "Firefighter"
        elif number == 8:
            return "Firefighter"

    # define a function to generate a random character name
    def random_character_name():
        # ask for a character name
        answer = input("Would you like to name your character? (y/n) ")
        # if the user answers yes, then ask for a character name
        if answer == "y":
            character_name = input("What is your character's name? ")
            return character_name
        # if the user answers no, then generate a random character name
        elif answer == "n":
            # generate a random number between 1 and 3
            number = random.randint(1, 8)
            # if the number is 1, then return "Bob"
            if number == 1:
                return "Jenny"
            # if the number is 2, then return "Joe"
            elif number == 2:
             return "Sasha"
            # if the number is 3, then return "Sally"
            elif number == 3:
                return "Kris"
            # if the number is 4, then return "Morgan"
            elif number == 4:
             return "Morgan"
            # if the number is 5, then return "Alex"
            elif number == 5:
                return "Alex"
            # if the number is 6, then return "Taylor"
            elif number == 6:
                return "Taylor"
            # if the number is 7, then return "Todd"
            elif number == 7:
                return "Todd"
            # if the number is 8, then return "Crystal"
            elif number == 8:
                return "Crystal"

    # define a function to generate a random character strength
    def random_character_strength():
        # generate a random number between 1 and 100
        number = random.randint(1, 100)
        # return the random number
        return number

    # define a function to generate a random character intellect
    def random_character_intellect():
        # generate a random number between 1 and 100
        number = random.randint(1, 100)
        # return the random number
        return number

    # define a function to generate a random character hit points
    def random_character_attractiveness():
        # generate a random number between 1 and 100
        number = random.randint(1, 100)
        # return the random number
        return number

    # define a function to generate a random character wisdom
    def random_character_wisdom():
        # generate a random number between 1 and 100
        number = random.randint(1, 100)
        # return the random number
        return number

    # define a function to generate a random character charisma
    def random_character_charisma():
        # generate a random number between 1 and 100
        number = random.randint(1, 100)
        # return the random number
        return number

    # define a function to generate a random character
    def random_character():
        # generate a random character type
        character_type = random_character_type()
        # generate a random character name
        character_name = random_character_name()
        # generate a random character strength
        character_strength = random_character_strength()
        # generate a random character intellect
        character_intellect = random_character_intellect()
        # generate a random character hit points
        character_attractiveness = random_character_attractiveness()
        # generate a random character wisdom
        character_wisdom = random_character_wisdom()
        # generate a random character charisma
        character_charisma = random_character_charisma()
        # return the random character
        return character_type, character_name, character_strength, character_intellect, character_attractiveness, character_wisdom, character_charisma

    # define a function to print the character information
    def print_character_info(character_type, character_name, character_strength, character_intellect, character_attractiveness, character_wisdom, character_charisma):
        # print the character information
        print("Character Type:", character_type)
        print("Character Name:", character_name)
        print("Character Strength:", character_strength, " out of 100")
        print("Character Intellect:", character_intellect, "out of 100")
        print("Character Attractiveness:", character_attractiveness, "out of 100")
        print("Character Wisdom:", character_wisdom, "out of 100")
        print("Character Charisma:", character_charisma, "out of 100")

    # define a function to ask the user if they want to create a character
    def ask_user():
        # ask the user if they want to create a character
        answer = input("Would you like to create a character? (y/n) ")
        # if the user answers yes, then return True
        if answer == "y":
            return True
        # if the user answers no, then return False
        elif answer == "n":
            return False

    # define a function to create a character
    def create_character():
        # generate a random character
        character_type, character_name, character_strength, character_intellect, character_attractiveness, character_wisdom, character_charisma = random_character()
        # print the character information
        print_character_info(character_type, character_name, character_strength, character_intellect, character_attractiveness, character_wisdom, character_charisma)

    # define a function to ask the user if they want to create another character
    def ask_user_again():
        # ask the user if they want to create another character
        answer = input("Would you like to create another character? (y/n) ")
        # if the user answers yes, then return True
        if answer == "y":
            return True
        # if the user answers no, then return False
        elif answer == "n":
            return False

    # define a function to run the program
    def run_program():
        # print the program title
        print("Random Character Generator")
        # print the program description
        print("----------------------------------")
        print("Welcome to the New Identity Hotline!")
        print("----------------------------------")
        print("Are you tired of your boring life and looking to spice things up?")
        print("----------------------------------")
        print("Do you want to be someone else for a day?")
        print("----------------------------------")
        print("We can create a new identity for you!")
        print("----------------------------------")
        # print a blank line
        print()

        # ask the user if they want to create a character
        answer = ask_user()
        # if the user answers yes, then run the program
        if answer == True:
            # create a character
            create_character()
            # ask the user if they want to create another character
            answer = ask_user_again()
            # if the user answers yes, then run the program
            if answer == True:
                # create a character
                print("----------------------------------")
                create_character()
                print("----------------------------------")
                print("Thank you for using the New Identity Hotline!")  
            # if the user answers no, then end the program
            elif answer == False:
                return
        # if the user answers no, then end the program
        elif answer == False:
            return

    # run the program
    run_program()

# run the program
main()
```
