```python
# import the sqlite3 database module
import sqlite3

# create a connection object that represents the database
conn = sqlite3.connect("myDatabase.db")

# create a cursor object that points to the database
cursor = conn.cursor()


# create a table called user
cursor.execute("CREATE TABLE IF NOT EXISTS user (first_name TEXT, last_name TEXT, email TEXT, phone_number TEXT, street_address TEXT, city TEXT, state TEXT, zipcode INTEGER);")

# Function of get user's first and last name
def getFirstLastName():
    # Store the results of the query into a list
    usernames = cursor.execute("SELECT first_name, last_name FROM user;")
    
    # Loop through the list to print the result
    for name in usernames:
        print(name[0], name[1])
      
# function to get user's name and contact
def getNameContact():
    # store the result of query into a list
    namecontact = cursor.execute("SELECT first_name, last_name, email, phone_number FROM user;")
    
    # Loop through the list to print the result
    for detail in namecontact:
        print(detail[0], detail[1], detail[2], detail[3])
        
# function to get user's name and address
def getNameAddress():
    # store the result of query into a list
    nameaddress = cursor.execute("SELECT first_name, last_name, street_address, city, state, zipcode FROM user;")
    
   # Loop through the list to print the result
    for detail in nameaddress:
        print(detail[0], detail[1], detail[2], detail[3], detail[4], detail[5])
  
# function to get all the details of the user
def getAll():
    # store the result of query into a list
    details = cursor.execute("SELECT first_name, last_name, email, phone_number, street_address, city, state, zipcode FROM user;")
   # Loop through the list to print the result
    for detail in details:
        print(detail[0], detail[1], detail[2], detail[3], detail[4], detail[5], detail[6], detail[7])

# ADD User, create a new function that asks inut for the following fields, first name, last name, email, phone number, street address, city, state, zipcode
def addUser():
    # get user input
    first_name = input("Please enter your first name: ")
    last_name = input("Please enter your last name: ")
    email = input("Please enter your email: ")
    phone_number = input("Please enter your phone number: ")
    street_address = input("Please enter your street address: ")
    city = input("Please enter your city: ")
    state = input("Please enter your state: ")
    zipcode = input("Please enter your zipcode: ")
    
    # insert the data into the table
    cursor.execute("INSERT INTO user VALUES (?,?,?,?,?,?,?,?);", (first_name, last_name, email, phone_number, street_address, city, state, zipcode))
    print("User added successfully!")

# main function
if __name__=="__main__":
    choice = 0
    while(1):
        print("""\nWelcome to the User Database!
              Please select the option:
                1 - Get All Details"
                2 - Get First and Last Name"
                3 - Get Name and Contact detail"
                4 - Get Name and Address detail"
                5 - Add User"
                0 - Exit""")
    
        choice = int(input())
        
        if choice == 0:
            print("Thank you for using the User Database!")
            print("Good Bye!")
            break
        if choice == 1:
            getAll()
        elif choice==2:
            getFirstLastName()
        elif choice==3:
            getNameContact()
        elif choice==4:
            getNameAddress()
        elif choice==5:
            addUser()
        else:
            print("Please Enter the right choice !\n")
                
    # commit the changes to the database
    conn.commit()
    
    # close the connection
    conn.close()
```
